# Step 1: Basic Models

In this section, we will explore three basic models: Linear, Polynomial Regression, and Spline. 

## 1.1 Generate Input Signal

Let's start by generating an input signal. We'll create different types of signals like sine, step, pulse, and noise.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import UnivariateSpline
from numpy.polynomial import Polynomial
import seaborn as sns

# Set global plot style
sns.set(style="darkgrid")

def generate_input_signal(signal_type='sine', duration=10, dt=0.01):
    t = np.arange(0, duration, dt)
    if signal_type == 'sine':
        return t, np.sin(2 * np.pi * t)
    elif signal_type == 'step':
        return t, np.ones_like(t)
    elif signal_type == 'pulse':
        pulse = np.zeros_like(t)
        pulse[int(len(t) / 4):int(len(t) / 4) + 50] = 1
        return t, pulse
    elif signal_type == 'noise':
        return t, np.random.normal(0, 1, len(t))
    else:
        return t, np.zeros_like(t)

# Example of generating a sine wave
t, signal = generate_input_signal('sine')
plt.plot(t, signal)
plt.title('Sine Wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

## 1.2 Linear Model

The linear model is a simple model where the output is a linear function of the input. The formula is:
$$ y(t) = a \cdot u(t) + b $$
where \( a \) is the slope and \( b \) is the intercept.

Let's visualize this model with some example parameters.


```python
def linear_model(u, a=1, b=0):
    return a * u + b

# Generate example input signal
t, u = generate_input_signal('sine')

# Apply linear model
a, b = 2, 1
y_linear = linear_model(u, a, b)

# Plot results
plt.figure(figsize=(10, 6))
plt.plot(t, u, label='Input Signal', linestyle='--', color='k')
plt.plot(t, y_linear, label='Linear Model', color='r')
plt.legend()
plt.title('Linear Model')
plt.xlabel('Time')
plt.ylabel('Output')
plt.show()
```

## 1.3 Polynomial Regression

Polynomial regression fits a polynomial to the input data. The degree of the polynomial can be adjusted to capture more complex patterns.

Lets fit a polynomial of degree 3 to our input signal.


```python
from numpy.polynomial import Polynomial

def polynomial_regression_model(u, degree=3):
    t = np.arange(len(u))
    p = Polynomial.fit(t, u, degree)
    return p(t)

# Apply polynomial regression model
degree = 3
y_poly = polynomial_regression_model(u, degree)

# Plot results
plt.figure(figsize=(10, 6))
plt.plot(t, u, label='Input Signal', linestyle='--', color='k')
plt.plot(t, y_poly, label=f'Polynomial Regression (Degree {degree})', color='b')
plt.legend()
plt.title('Polynomial Regression Model')
plt.xlabel('Time')
plt.ylabel('Output')
plt.show()
```

## 1.4 Spline Model

Spline interpolation provides a smooth curve through the data points. The smoothness parameter can be adjusted to control the degree of smoothness.

Lets apply spline interpolation to our signal.


```python
from scipy.interpolate import UnivariateSpline

def spline_model(u, smoothing=0):
    t = np.arange(len(u))
    spline = UnivariateSpline(t, u, s=smoothing)
    return spline(t)

# Apply spline model
smoothing = 0.5
y_spline = spline_model(u, smoothing)

# Plot results
plt.figure(figsize=(10, 6))
plt.plot(t, u, label='Input Signal', linestyle='--', color='k')
plt.plot(t, y_spline, label=f'Spline Model (Smoothing {smoothing})', color='orange')
plt.legend()
plt.title('Spline Model')
plt.xlabel('Time')
plt.ylabel('Output')
plt.show()
```
